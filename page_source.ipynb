```python
import time
from selenium import webdriver
```

```python
driver = webdriver.Chrome()

url = "https://www.gist.ac.kr/kr/html/sub05/050602.html?mode=V&no=215761"
driver.get(url)

time.sleep(3)
```

```python
html_source = driver.page_source
print(html_source)

with open('page_source.html', 'w', encoding='utf-8') as f:
    f.write(html_source)
```
